```python
# (2) выведите, что за типы переменных, сколько пропусков,
# для численных значений посчитайте пару статистик (в свободной форме)
for col in df.columns: 
    column_type = str(type(df.loc[0, col]))
    print('Поле: ' + col + ', тип: ' + column_type + ', к-во NaN:', len(df[df[col].isnull()]), end='')
    if (("int" in column_type) | ("float" in column_type)):
        print(', к-во нулей:', len(df[df[col]==0]),', min=', df[col].min(), ', max=', df[col].max(), ', mean=', df[col].mean())
    else: 
        print(', к-во пустых строк:', len(df[df[col]=='']))
```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000
df.loc[df['LIMIT_BAL']<=10000,'new_col']='A'
df.loc[(df['LIMIT_BAL']<=100000)&(df['LIMIT_BAL']>10000),'new_col']='B'
df.loc[(df['LIMIT_BAL']<=200000)&(df['LIMIT_BAL']>100000),'new_col']='C'
df.loc[(df['LIMIT_BAL']<=400000)&(df['LIMIT_BAL']>200000),'new_col']='D'
df.loc[(df['LIMIT_BAL']<=700000)&(df['LIMIT_BAL']>400000),'new_col']='E'
df.loc[df['LIMIT_BAL']>700000,'new_col']='F'
df['new_col'].unique()
#TODO
```

```python
# (10) построить большой график (подсказка - используя seaborn) для построения завимисости всех возможных пар параметров
# разным цветом выделить разные значение "default payment next month"
# (но так как столбцов много - картинка может получиться "монструозной")
# (поэкспериментируйте над тем как построить подобное сравнение параметров)
# (подсказка - ответ может состоять из несколькольких графиков)
# (если не выйдет - программа минимум - построить один график со всеми параметрами)
import seaborn as sns
columns = df.columns
col = [columns[~((columns.str.contains('PAY_'))|(columns.str.contains('BILL_')))],
       columns[(columns.str.contains('PAY_'))&(~columns.str.contains('PAY_AMT'))|(columns=='default.payment.next.month')],
       columns[columns.str.contains('PAY_AMT')|(columns=='default.payment.next.month')],
       columns[columns.str.contains('BILL_')|(columns=='default.payment.next.month')]
      ]
for i in range(4): sns.pairplot(data=df[col[i]], hue='default.payment.next.month', size=2.5, aspect=1)
#TODO
```

```python
nmax=5
_, axes = plt.subplots(min(nmax,(len(columns)-2)**2),figsize=(23,50))
n=0
for i,icol in enumerate(columns):
    for j,jcol in enumerate(columns): 
        if n==nmax: break
        if (i==j)|(icol=='default.payment.next.month')|(icol=='ID')|(jcol=='default.payment.next.month'): continue
        print(n,i,icol,j,jcol,)
        sns.factorplot(x=icol,y=jcol,data=df, hue='default.payment.next.month', kind='bar', ax=axes[n])
        n+=1
```
